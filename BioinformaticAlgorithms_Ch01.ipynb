```python
## Compute the Number of Times a Pattern Appears in a Text
# rosalind_ba1a
def countPatternInText(text,pattern):
  tSize=len(text)
  pSize=len(pattern)
  if pSize > tSize:
    #raise Exception('pattern is larger than text')  
    print('no')
  else:
    counter=0
    for i in range(0,tSize-pSize+1):
      if text[i:i+pSize] == pattern:
        counter += 1
    return counter
    
```

```python
def hashPatternInText(text,pattern):
  kmerDict = {}
  tSize=len(text)
  pSize=len(pattern)
  for i in range(0,tSize-pSize+1):
        kmer = text[i:i+pSize]
        if kmer not in kmerDict.keys():
          kmerDict[kmer] = 1
        else:
          kmerDict[kmer] += 1
  return kmerDict.get(pattern)
```

#### BA1B
http://rosalind.info/problems/ba1b/

We say that Pattern is a most frequent k-mer in Text if it maximizes Count(Text, Pattern) among all k-mers. 
For example, "ACTAT" is a most frequent 5-mer in "ACAACTATGCATCACTATCGGGAACTATCCT", and "ATA" is a most frequent 3-mer of "CGATATATCCATAG".

```python
import pandas as pd
def hashKmers(text,k):
  kmerDict = {}
  tSize=len(text)
  kSize=k
  for i in range(0,tSize-kSize+1):
        kmer = text[i:i+kSize]
        if kmer not in kmerDict.keys():
          kmerDict[kmer] = 1
        else:
          kmerDict[kmer] += 1
  allK = pd.DataFrame([kmerDict],index=['n']).T
  return list(allK[allK.n == max(allK.n)].index)
```

#### BA1D Find All Occurrences of a Pattern in a String
http://rosalind.info/problems/ba1d/

In this problem, we ask a simple question: how many times can one string occur as a substring of another? Recall from “Find the Most Frequent Words in a String” that different occurrences of a substring can overlap with each other. For example, ATA occurs three times in CGATATATCCATAG.
Pattern Matching Problem

**Find all occurrences of a pattern in a string.** <br>
Given: Strings Pattern and Genome. <br>
Return: All starting positions in Genome where Pattern appears as a substring. Use 0-based indexing.<br>

```python
def hashPatternInText(text,pattern):
  kmerDict = {}
  tSize=len(text)
  pSize=len(pattern)
  for i in range(0,tSize-pSize+1):
        kmer = text[i:i+pSize]
        if kmer not in kmerDict.keys():
          kmerDict[kmer] = [i]
        else:
          kmerDict[kmer].append(i)
  return kmerDict.get(pattern)
```

```python
def hashPatternInText(text,k):
  kmerDict = {}
  tSize=len(text)
  pSize=k
  for i in range(0,tSize-pSize+1):
        kmer = text[i:i+pSize]
        if kmer not in kmerDict.keys():
          kmerDict[kmer] = [i]
        else:
          kmerDict[kmer].append(i)
  return kmerDict
```

```python
def getClumps(dictKmers,l,t):
    kmerClumps = []
    for k,v in dictKmers.items():
        if len(v) == t:
            if ( max(v)-min(v) ) <= l:
                kmerClumps.append(k)
    return kmerClumps
```

#### BA1F Find a Position in a Genome Minimizing the Skew
http://rosalind.info/problems/ba1f/

Define the skew of a DNA string Genome, denoted Skew(Genome), as the difference between the total number of occurrences of 'G' and 'C' in Genome. Let Prefixi (Genome) denote the prefix (i.e., initial substring) of Genome of length i. For example, the values of Skew(Prefixi ("CATGGGCATCGGCCATACGCC")) are:

_C  A  T  G G G C A T C G G C C A T A C  G C C_ <br>
  ..  0 -1 -1 -1 0 1 2 1 1 1 0 1 2 1 0 0 0 0 -1 0 -1 -2

**Minimum Skew Problem**

_Find a position in a genome minimizing the skew._ <br>
Given: A DNA string Genome. <br>

Return: All integer(s) i minimizing Skew(Prefixi (Text)) over all values of i (from 0 to |Genome|). <br>

```python
def dict2pd(myDict,colname):
    return pd.DataFrame([myDict],index=[colname]).T

import pandas as pd
def minSkew(text):
    costDict = {'A':0,'C':-1,'T':0,'G':1}
    runSum = {} #Start counter
    counter = 0
    # +1 if g, -1 if c
    for i,nt in enumerate(text):
        counter = counter + costDict[nt]
        runSum[i] = counter 
    runSum = dict2pd(runSum,'pos')
    minCumSum = list(runSum[runSum.pos == min(runSum.pos)].index)
    return ' '.join( [str(_ + 1) for _ in minCumSum])
```

#### BA1G Compute the Hamming Distance Between Two Strings
http://rosalind.info/problems/ba1g/

We say that position i in k-mers p1 … pk and q1 … qk is a mismatch if pi ≠ qi. For example, CGAAT and CGGAC have two mismatches. The number of mismatches between strings p and q is called the Hamming distance between these strings and is denoted HammingDistance(p, q).

**Hamming Distance Problem**<br>

_Compute the Hamming distance between two DNA strings._ <br>

Given: Two DNA strings.<br>
Return: An integer value representing the Hamming distance.<br>

```python
def hammingDistance(s1,s2):
    if len(s1) != len(s2):
        raise Exception ('strings different sizes')
    else:
        hd = 0
        for i,v in enumerate(s1):
            if s1[i] != s2[i]:
                hd += 1
        return hd
```

#### BA1H Find All Approximate Occurrences of a Pattern in a String 
http://rosalind.info/problems/ba1h/


We say that a k-mer Pattern appears as a substring of Text with at most d mismatches if there is some k-mer substring Pattern' of Text having d or fewer mismatches with Pattern, i.e., HammingDistance(Pattern, Pattern') ≤ d. Our observation that a DnaA box may appear with slight variations leads to the following generalization of the Pattern Matching Problem.

Approximate Pattern Matching Problem
Find all approximate occurrences of a pattern in a string.

Given: Strings Pattern and Text along with an integer d.

Return: All starting positions where Pattern appears as a substring of Text with at most d mismatches.

```python
def patternInText_approx(text,pattern,minDist):
  pos = []
  tSize=len(text)
  pSize=len(pattern)
  for i in range(0,tSize-pSize+1):
        kmer = text[i:i+pSize]
        if hammingDistance(kmer,pattern) <=minDist: 
          pos.append(i)
  return pos
```

#### BA1I Find the Most Frequent Words with Mismatches in a String
http://rosalind.info/problems/ba1i/


We defined a mismatch in “Compute the Hamming Distance Between Two Strings”. We now generalize “Find the Most Frequent Words in a String” to incorporate mismatches as well.

Given strings Text and Pattern as well as an integer d, we define Countd(Text, Pattern) as the total number of occurrences of Pattern in Text with at most d mismatches. For example, Count1(AACAAGCTGATAAACATTTAAAGAG, AAAAA) = 4 because AAAAA appears four times in this string with at most one mismatch: AACAA, ATAAA, AAACA, and AAAGA. Note that two of these occurrences overlap.

A most frequent k-mer with up to d mismatches in Text is simply a string Pattern maximizing Countd(Text, Pattern) among all k-mers. Note that Pattern does not need to actually appear as a substring of Text; for example, AAAAA is the most frequent 5-mer with 1 mismatch in AACAAGCTGATAAACATTTAAAGAG, even though AAAAA does not appear exactly in this string. Keep this in mind while solving the following problem.

**Frequent Words with Mismatches Problem**<br>

__Find the most frequent k-mers with mismatches in a string.__ <br>

Given: A string Text as well as integers k and d.<br>
Return: All most frequent k-mers with up to d mismatches in Text.<br>

```python
import itertools
import pandas as pd
```

```python
#### Create all possible k permutations of nucleotides
joinNT = lambda x: [str(_)+x for _ in nt]
nt=['A','C','T','G']
def kPermutations(i):
    tmp=nt
    while i>1:
        tmp=[joinNT(_) for _ in tmp]
        tmp=sum(tmp, [])
        i -= 1
    return tmp
```

```python
#### kmerize genome
def kmerize(text,kSize):
  kmers = []
  tSize=len(text)
  for i in range(0,tSize-kSize+1):
        kmers.append(text[i:i+kSize])
  return kmers
```

```python
def freqKmers_mismatch(s,k,d):
    kmerList = kmerize(s,k) # Kmerize the string
    allK = kPermutations(k) # Get all possible kmers of k size
    ## For each kmer in the string (k in allK): 
    # check similar kmers (by hamming distance) and add a count if they are similar 
    # result is a matrix of frequency including similarities 
    mm = pd.DataFrame([[1 if hammingDistance(k,_) <= d else 0 for _ in kmerList] for k in allK ],index=allK,columns=kmerList)
    # sum across columns
    mm = pd.DataFrame(mm.sum(1),columns=['count']).sort_values(by='count',ascending=False)
    return list(mm[mm['count']==max(mm['count'])].index), mm
```

#### BA1L Implement PatternToNumber
http://rosalind.info/problems/ba1l/


**Implement PatternToNumber** <br>

_Convert a DNA string to a number._ <br>

Given: A DNA string Pattern.<br>
Return: PatternToNumber(Pattern).<br>

```python
## Recursive function that calculates the lexicographic position of a kmer of size k 
nt2idx = {'A':0,'C':1,'G':2,'T':3} #Symbol to nucleotide
def P2N(pattern):
    k=len(pattern)
    if k == 0: #Base case
        return 0
    else:
        symbol = pattern[-1]
        return 4 * P2N(pattern[0:k-1]) + nt2idx[symbol]
```

```python
idx2nt = {0:'A',1:'C',2:'G',3:'T'} #Symbol to nucleotide
def NumberToPattern(i,k):
    if k==1:
        return idx2nt[i]
    else:
        prefix = i//4
        remainder = i%4
        symbol = idx2nt[remainder]
        prefPattern = NumberToPattern(prefix,k-1)
        return  prefPattern + symbol
    
```

#### BA1N Generate the d-Neighborhood of a String
http://rosalind.info/problems/ba1n/

The d-neighborhood Neighbors(Pattern, d) is the set of all k-mers whose Hamming distance from Pattern does not exceed d.

**Generate the d-Neighborhood of a String** <br>
_Find all the neighbors of a pattern._ <br>

Given: A DNA string Pattern and an integer d. <br>
Return: The collection of strings Neighbors(Pattern, d). <br>

```python
pattern='ATG'
k = len(pattern)
k
```

```python
def dNeighbors(pattern,d):
    dNeigh=[]
    k = len(pattern)
    for _ in range(0,(4**k)):
        kmer = NumberToPattern(_,k)
        if hammingDistance(kmer,pattern) <= d:
            dNeigh.append(kmer)
    return "\n".join(dNeigh)
```
